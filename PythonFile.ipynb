```python
import statistics
import pandas as pd
import numpy as np
import tkinter as tk
import matplotlib.pyplot as plt
from matplotlib import animation
from matplotlib.figure import Figure
from matplotlib.backends.backend_tkagg import (FigureCanvasTkAgg, 
NavigationToolbar2Tk)
from matplotlib.backends.backend_pdf import PdfPages
from pandas.plotting import table
from tkinter import *
from tkinter import ttk
global ani  #animation
global ani2
global zoom
global spec #spectogram
global GraphNamesList
global ax    #axis
global speed

pause = False

zoom =0

spec=0

table_df=pd.DataFrame()
GraphNamesList = []

X=0
Y = 0
Z = 0

speed = 100
show = True
show2 = True
show3 = True
color = 'b'
color2= 'g'
color3= 'y'

colorpa1 = 'rainbow'  #color palate
colorpa2 = 'rainbow'
colorpa3 = 'rainbow'
contrastmin = -100
contrastmax = -20

#statstics calculation function
def CalculateData(x,y):
    mean = y.mean()
    minimum = min(y)
    maximum = max(y)
    duration = len(x)
    result = {'mean':mean,'min':minimum,'max':maximum,'duration':duration}
    return result
#Spectogram contrast function
def changecontrast():
    global contrastmin
    global contrastmax
    contrastmax = Slider4.get()
    contrastmin =  Slider3.get()
def changepal(self):
    global colorpa1
    colorpa1= colorpa1_box.get()
    spec1()
def changepal2(self):
    global colorpa2
    colorpa2= colorpa2_box.get()
    spec2()
def changepal3(self):
    global colorpa3
    colorpa3= colorpa3_box.get() 
    spec3()


def ZoomIn():
    global zoom
    zoom+=1
    
def ZoomOut():
    global zoom
    zoom-=1
    


def hide1():
    global show
    show = False
def show1():
    global show
    show = True
def hide2():
     global show2
     show2 = False
def show2():
    global show2
    show2 = True
def hide3():
    global show3
    show3 = False
def show3():
    global show3
    show3 = True    
    
def ChangeColor(event):
    global color 
    color= ColorCB.get() 
    
def ChangeColor2(event):
    global color2 
    color2= ColorCB2.get() 
    
def ChangeColor3(event):
    global color3 
    color3= ColorCB3.get() 

    
def faster():
        global speed   
        Pause()
        if speed >99:
            speed = speed - 99
        ani.event_source.interval = speed
        Resume()
       


        
        
def slower():
        global speed  
        Pause()
        speed = speed + 99
        ani.event_source.interval = speed
        Resume()


        

def Pause():
        global pause
        pause = True  
        
def Resume():
        global pause
        pause = False
def openFile():
    filepath = filedialog.askopenfilename(filetypes= (("csv files",".csv"),("all files",".*")))
    openFile.filepath = filepath
    file = open(filepath,'r')
    print(filepath)
    graph1()
    file.close()
  
def openFile2():
    filepath = filedialog.askopenfilename(filetypes= (("csv files",".csv"),("all files",".*")))
    openFile.filepath = filepath
    file = open(filepath,'r')
    print(filepath)
    graph2()
    file.close()
 # creating the Tkinter canvas
    # containing the Matplotlib figure

def openFile3():
    filepath = filedialog.askopenfilename(filetypes= (("csv files",".csv"),("all files",".*")))
    openFile.filepath = filepath
    file = open(filepath,'r')
    print(filepath)
    graph3()
    file.close()    


 # creating the Tkinter canvas
    # containing the Matplotlib figure

def graph1():
        global hide
        global ani
        global fig
        global ax1 #graph axis
        global ax2 #spectogram axis
        global volt1
        global sampling
        global Slider1
        global Slider2
        global fig2
        
        loaded_data = pd.read_csv( openFile.filepath)
        time = loaded_data["time_sec"] 
        volt1 = loaded_data["amplitude"]
        sampling = 300
        max_y=max(volt1)
        min_y=min(volt1)
        Slider1 = Scale(frame2, from_=max_y, to=min_y,resolution=( abs(max_y) + abs(min_y) )/100)
        Slider1.set(0)
        Slider1.pack(side = LEFT)
        Slider2 = Scale(frame2, from_=0, to=len(volt1)-101, orient=HORIZONTAL)
        Slider2.set(0)
        Slider2.pack(side = LEFT)


       
        fig,(ax1) = plt.subplots(1)
        fig2,(ax2) = plt.subplots(1)
        
        label1 = simpledialog.askstring(title="Signal1",
                                prompt="Enter Signal1 Name:")

        #Button for adding spectofram
        button10=tk.Button(frame_bottom,text=label1,command = spec1)
        button10.pack(side = LEFT)
       
        canvas = FigureCanvasTkAgg(fig,
                                   master =m)        
        # placing the canvas on the Tkinter window
                   
        m.add(canvas.get_tk_widget())
        canvas2 = FigureCanvasTkAgg(fig2,
                                   master =m)  
        canvas2.draw()
      
        # placing the canvas on the Tkinter window
                   
        m.add(canvas2.get_tk_widget())
      
                    

                      
                      
       
         
      
       
        global text1        
        global table_df
        text1=ax1.text(0,0,"")
        GraphNamesList.append(label1)
        row1=CalculateData(time,volt1) #row in the data frame
        table_df=table_df.append(row1,ignore_index=True)
        table_df['Graph Names']= GraphNamesList
        table_df=table_df.set_index('Graph Names')
        

       
        def animate(i):  
            global X  #counter for pause & resume operations
            global text1
            if show:
                ax1.plot(time, volt1,color)
            else:
                ax1.plot(time,volt1,'w',linewidth=2.0)
            X=Slider2.get()
            ax1.set_xlim(time[Slider2.get()]+(zoom/100),time[Slider2.get()+100]-(zoom/100))
            ax1.set_ylim((min(volt1)+(zoom/10)+Slider1.get()),max(volt1)-(zoom/10)+Slider1.get())
            xmin, xmax = ax1.get_xlim()
            label_x= xmin+((xmax-xmin)/4)
            label_y=volt1[ (((Slider2.get()+100)-(Slider2.get()))/4)+(Slider2.get())] 
            m.config(orient=HORIZONTAL)

            if show:
                text1.set_visible(False)
                text1=ax1.text(label_x,label_y,label1)
                text1.set_visible(True)
            else:
                text1.set_visible(False)
                

            if not pause: 
                X=X+1
                Slider2.set(X)
                if Slider2.get()>=len(volt1)-101:
                    Slider2.set(0)
            changecontrast()
            if spec==1: 
                spec1()
            elif spec==2:
                spec2()
            elif spec==3: 
                spec3()
                

        
        
            
    
                    
            
        
        ani = animation.FuncAnimation(fig, animate, interval = speed,frames =len(volt1)-100,blit=False)
def graph2():
    
        global ani2
        global volt2
        loaded_data = pd.read_csv( openFile.filepath)
        label2 = simpledialog.askstring(title="Signal2",
                                prompt="Enter Signal2 Name:")
        
        button11=tk.Button(frame_bottom,text=label2,command = spec2)
        button11.pack(side = LEFT)
        time = loaded_data["time_sec"] 
        volt2 = loaded_data["amplitude"]
        ax1.set_ylim(min(volt1+volt2),max(volt1+volt2))



        # creating the Matplotlib toolbar
       
       
        global text2
        text2=ax1.text(0,0,"")
        global table_df
       
        GraphNamesList.append(label2)
        row2=CalculateData(time,volt2)
        table_df=table_df.append(row2,ignore_index=True)
        table_df['Graph Names']= GraphNamesList
        table_df=table_df.set_index('Graph Names')
        def animate2(i):  
            global Y  
            global text2
            if show2:
                ax1.plot(time, volt2,color)
            else:
                ax1.plot(time,volt2,'w',linewidth=2.0)
            xmin, xmax = ax1.get_xlim()
            label_x2= xmin+((xmax-xmin)/2)
            label_y2=volt2[ (((Slider2.get()+100)-(Slider2.get()))/2)+(Slider2.get())]
            if show2:
                text2.set_visible(False)
                text2=ax1.text(label_x2,label_y2,label2)
                text2.set_visible(True)
            else:
                text2.set_visible(False)

            
            if not pause: 
                Y=Y+1
                    
        
            
    
                    
            
        
        ani2 = animation.FuncAnimation(fig, animate2, interval = speed,frames =len(volt2)-100,blit=False)         
def graph3():
    
        global ani3
        global volt3
        loaded_data = pd.read_csv( openFile.filepath)
        label3 = simpledialog.askstring(title="Signal3",
                                prompt="Enter Signal3 Name:")        
        
        button12=tk.Button(frame_bottom,text=label3,command = spec3)
        button12.pack(side = LEFT)
        time = loaded_data["time_sec"] 
        volt3 = loaded_data["amplitude"]
        ax1.set_ylim(min(volt1+volt2+volt3),max(volt1+volt2+volt3))



        # creating the Matplotlib toolbar
       
       
        global text3
        text3=ax1.text(0,0,"")
        global table_df
        GraphNamesList.append(label3)
        row3=CalculateData(time,volt3)
        table_df=table_df.append(row3,ignore_index=True)
        table_df['Graph Names']= GraphNamesList
        table_df=table_df.set_index('Graph Names')
        def animate3(i):  
            global Z
            global text3
            if show3:
                ax1.plot(time, volt3,color)
            else:
                ax1.plot(time,volt3,'w',linewidth=2.0)
            xmin, xmax = ax1.get_xlim()
            label_x3= xmin+((xmax-xmin)*3/4)
            label_y3=volt3[ (((Slider2.get()+100)-(Slider2.get()))*3/4)+(Slider2.get())]
            if show3:
                text3.set_visible(False)
                text3=ax1.text(label_x3,label_y3,label3)
                text3.set_visible(True)
            else:
                text3.set_visible(False)

            
            if not pause: 
                Z=Z+1
                    
        
            
    
                    
            
        
        ani3 = animation.FuncAnimation(fig, animate3, interval = speed,frames =len(volt3)-100,blit=False)
def spec1():
            global spec
            spec=1
            ax2.clear()
            ax2.plot = ax2.specgram(volt1,Fs=sampling,cmap = colorpa1,vmin = contrastmin,vmax =contrastmax)  

def spec2():
            global spec
            spec =2
            ax2.clear()
            ax2.plot = ax2.specgram(volt2,Fs=sampling,cmap = colorpa2,vmin = contrastmin,vmax =contrastmax)
            
         
def spec3():
            global spec
            spec =3
            ax2.clear()
            ax2.plot = ax2.specgram(volt3,Fs=sampling,cmap = colorpa3,vmin = contrastmin,vmax =contrastmax)

def ExportToPDF():
    global table_df

    with PdfPages('test2_pdf.pdf') as pdf:
        pdf.savefig(fig)
        pdf.savefig(fig2)
        fig3=plt.figure()
        ax3 = plt.subplot(111, frame_on=False)
        ax3.xaxis.set_visible(0)
        ax3.yaxis.set_visible(0)
        table(ax3, table_df, loc='upper center')
        pdf.savefig(fig3)
    #plt.savefig('SignalViwer.pdf')
    #fig.savefig('sing.pdf')
    # pp = PdfPages('foo.pdf')
    # pp.savefig(table)
    #pp.savefig(fig)
    #pp.savefig()
    #pp.close()          
        
root = tk.Tk()
frame_bottom = tk.Frame(root)
frame_bottom.pack(expand = True,fill = 'both')
frame2 = tk.Frame(root)
frame2.pack(side = BOTTOM)
m = PanedWindow(root,bd =6)
m.pack(fill=BOTH, expand=True)


SelectedColor = tk.StringVar()
red = 'r'
blue='b'
green = 'g'
yellow = 'y'
Colors = ['red','blue','green','yellow'] 


ColorCB = ttk.Combobox(frame_bottom, textvariable=SelectedColor)
ColorCB['values'] = Colors
ColorCB['state'] = 'readonly'  # normal
ColorCB.pack(side = LEFT)



ColorCB.bind('<<ComboboxSelected>>', ChangeColor)



colorpalette1 = tk.StringVar()
colorpalette2 = tk.StringVar()
colorpalette3 = tk.StringVar()
colorpalette1 = ['coolwarm','magma','copper','plasma','twilight']
colorpalette2 = ['coolwarm','magma','copper','plasma','twilight']
colorpalette3 = ['coolwarm','magma','copper','plasma','twilight']
colorpa1_box = ttk.Combobox(frame2, textvariable=colorpalette1)
colorpa2_box = ttk.Combobox(frame2, textvariable=colorpalette2)
colorpa3_box = ttk.Combobox(frame2, textvariable=colorpalette3)
colorpa1_box['values'] = colorpalette1
colorpa2_box['values'] = colorpalette2
colorpa3_box['values'] = colorpalette3
colorpa1_box['state'] = 'readonly'
colorpa2_box['state'] = 'readonly'
colorpa3_box['state'] = 'readonly'
colorpa1_box.pack(side = LEFT)
colorpa2_box.pack(side = LEFT)
colorpa3_box.pack(side = LEFT)
colorpa1_box.bind('<<ComboboxSelected>>',changepal)
colorpa2_box.bind('<<ComboboxSelected>>',changepal2)
colorpa3_box.bind('<<ComboboxSelected>>',changepal3)


frame2 = tk.Frame(root)
frame2.pack(side = BOTTOM)

#open files buttons
button = tk.Button(frame_bottom,text="ch1",command=openFile)
button2 = tk.Button(frame_bottom,text="ch2",command=openFile2)
button3 = tk.Button(frame_bottom,text="ch3",command=openFile3)
#Show & Hide buttons
button4 = tk.Button(frame_bottom,text = "hide ch1",command = hide1)
button5 = tk.Button(frame_bottom,text = "show ch1",command = show1)
button6 = tk.Button(frame_bottom,text = "hide ch2",command = hide2)
button7 = tk.Button(frame_bottom,text = "show ch2",command = show2)
button8 = tk.Button(frame_bottom,text = "hide ch3",command = hide3)
button9 = tk.Button(frame_bottom,text = "show ch3",command = show3)
#Zoom Buttons
button13=tk.Button(frame_bottom,text="ZoomIn",command = ZoomIn)
button14=tk.Button(frame_bottom,text="ZoomOut",command = ZoomOut)
#export Buttons
button15=tk.Button(frame_bottom,text="Export",command = ExportToPDF)

#. pack funtion is to add things on the window
button.pack(side = LEFT)
button2.pack(side = LEFT)
button3.pack(side = LEFT)
pausebutton = tk.Button(frame_bottom,text="pause",command=Pause)
playbutton = tk.Button(frame_bottom,text="play",command=Resume)
fasterbutton = tk.Button(frame_bottom,text = "faster",command=faster)
slowerbutton = tk.Button(frame_bottom,text = "slower",command=slower)
fasterbutton.pack(side = LEFT)
slowerbutton.pack(side = LEFT)
pausebutton.pack(side = LEFT)
playbutton.pack(side = LEFT)
button4.pack(side = LEFT)
button5.pack(side = LEFT)
button6.pack(side = LEFT)
button7.pack(side = LEFT)
button8.pack(side = LEFT)
button9.pack(side = LEFT)
button13.pack(side = LEFT)
button14.pack(side = LEFT)
button15.pack()

root.title('Signal Viewer')
Slider3 = Scale(frame2,from_=-100,to=-50)
Slider3.set(3000)
Slider3.pack(side = RIGHT)
Slider4 = Scale(frame2,from_=-50,to=-20)
Slider4.set(5000)
Slider4.pack(side = RIGHT)
root.update()


root.mainloop()
```
